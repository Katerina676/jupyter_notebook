```python
cols = ['Equipment', 'Age', 'BodyweightKg', 'WeightClassKg', 'Best3SquatKg', 'Best3BenchKg','Best3DeadliftKg', 'Place' , 'TotalKg', 'Sex']
data = data[cols]
data.head()
```

```python
fig = plt.figure(figsize=(14,5))
data['Equipment'].value_counts().plot.bar()
```

```python
fig = plt.figure(figsize=(14,5))
data['Equipment'].value_counts().sort_index().plot.bar()
```

```python
fig = plt.figure(figsize=(14,5))
(data['Equipment'].value_counts() / len(data)).plot.bar()
```

```python
data.groupby(['Equipment'])[['BodyweightKg', 'TotalKg']].mean()
```

```python
data.groupby(['Equipment'])[['BodyweightKg', 'TotalKg']].mean().plot.bar(figsize=(14,5), stacked=True)
```

```python
fig = plt.figure(figsize=(14,5))
chart = data['Equipment'].value_counts().plot.line()
plt.show()
```

```python
fig = plt.figure(figsize=(14,5))
chart = data['Equipment'].value_counts().plot.area()
plt.show()
```

```python
fig = plt.figure(figsize=(14,5))
data['Age'].plot.hist(bins=10)
```

```python
fig = plt.figure(figsize=(14,5))
data['BodyweightKg'].plot.hist(bins=2)
```

```python
data[data['Age']<50].sample(300).plot.scatter(x='BodyweightKg', y='Age', figsize=(14,5))
```

```python
data[data['Age']<50].plot.scatter(x='BodyweightKg', y='Age', figsize=(14,5))
```

```python
data[data['Age']<50].plot.hexbin(x='BodyweightKg', y='Age', figsize=(17,6), gridsize = 25, cmap = 'Greens')
```

```python
fig = plt.figure(figsize=(14,5))
sns.countplot(data['Equipment'])
```

```python
fig = plt.figure(figsize=(20,10))
axl = fig.add_subplot(221)
sns.kdeplot(data.Age)
ax2 = fig.add_subplot(222)
data['Age'].value_counts().sort_index().plot.line()
```

```python
fig = plt.figure(figsize=(14,5))
sns.kdeplot(data.loc[:,['BodyweightKg', 'Age']].dropna().sample(5000), color='r', cmap="Reds", shade=True, shade_lowest=False)
```

```python
fig = plt.figure(figsize=(14,5))
sns.distplot(data['Age'].dropna(),kde=False)
```

```python
sns.jointplot(x=data['BodyweightKg'], y=data['Age'], kind='hex', gridsize=20)
```

```python
sns.boxplot(y="Age", x="BodyweightKg", data=data[data.Age.isin(np.arange(20,40,1))], orient="h")
```

```python
fig = plt.figure(figsize=(14,5))
sns.violinplot(y="TotalKg", x="Equipment", hue='Sex', split=True, data=data)
```
